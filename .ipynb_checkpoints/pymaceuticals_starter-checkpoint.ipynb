```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
study_data_complete = pd.merge(study_results, mouse_metadata, how="left", on="Mouse ID")

# Display the data table for preview
study_data_complete.head()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

# series variable to hold Tumor Volume Data grouped by Drug Regimen

# variable to hold the Mean Tumor Volume Data Grouped by Drug Regimen
MeanTumorValue=study_data_complete_mod.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()

# variable to hold median Tumor Volume Data Grouped by Drug Regimen
MedianTumorValue=study_data_complete_mod.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()

# variable to hold the Tumor Volume Variance Data Grouped by Drug Regimen
TumorValueVar=study_data_complete_mod.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()

# variable to hold the Tumor Volume Standard Deviation Data Grouped by Drug Regimen
TumorValueStd=study_data_complete_mod.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()

# variable to hold the Tumor Volume SEM Data Grouped by Drug Regimen
TumorValueSem=study_data_complete_mod.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()

# Convert to DataFrame
SummaryTable=pd.DataFrame({"Mean Tumor Volume":MeanTumorValue,
                          "Median Tumor Volume":MedianTumorValue,
                          "Tumor Volume Variance":TumorValueVar,
                          "Tumor Volume Stdev":TumorValueStd,
                          "Tumor Volume SEM":TumorValueSem
                          })

#Depending on which duplicate values mouse id & timepoint for mice g989 are dropped/kept
#(i.e First duplicate occurence or last duplicate occurence or none or someother criteria)
#Values for Propriva Drug Regimen will change, as the mice g989 was given Propriva Drug Regimen

# Preview DataFrame
SummaryTable
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line

study_data_complete_mod.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg([ # Group by Drug Regimen
    ('Mean_Tumor_Volume','mean'), #Output Column name,Name of Function to be applied on Tumor Volume (mm3)
    ('Median_Tumor_Volume','median'),
    ('Variance_Tumor_Volume','var'),
    ('Stdev_Tumor_Volume','std'),
    ('SEM_Tumor_Volume','sem')])

#Depending on which duplicate values mouse id & timepoint for mice g989 are dropped/kept
#(i.e First duplicate occurence or last duplicate occurence or none or someother criteria)
#Values for Propriva Drug Regimen will change, as the mice g989 was given Propriva Drug Regimen
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
TragetDrugs=['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# set drug regimen as index and drop associated regimens while only keeping Capomulin, Ramicane, Infubinol, and Ceftamin
TargetDrugTumorVolume=study_data_complete_mod[study_data_complete_mod['Drug Regimen'].isin(TragetDrugs)]

# isolated view of just capomulin for later use
CapomulinData=TargetDrugTumorVolume[TargetDrugTumorVolume['Drug Regimen']=='Capomulin']

# Reset index so drug regimen column persists after inner merge
TargetDrugTumorVolume.reset_index()

# get mouse count per drug
MouseCountDF=pd.DataFrame(TargetDrugTumorVolume.groupby('Drug Regimen')['Mouse ID'].agg([('Unique Mouse Count','nunique')]))

# Start by getting the last (greatest) timepoint for each mouse
MouseMaxTimePoint=TargetDrugTumorVolume.groupby('Mouse ID')['Timepoint'].max()
MouseMaxTimePoint.head()
```

```python
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
FinalTumorVolData=pd.merge(MouseMaxTimePoint,TargetDrugTumorVolume,on=['Mouse ID','Timepoint'])
#TargetDrugTumorVolume.merge(MouseMaxTimePoint,on=['Mouse ID','Timepoint']).head()
# show all rows of data
FinalTumorVolData=FinalTumorVolData[['Mouse ID','Drug Regimen','Timepoint','Tumor Volume (mm3)','Metastatic Sites','Sex','Age_months','Weight (g)']]
pd.set_option('display.max_rows',100)
FinalTumorVolData.set_index('Mouse ID')
```

```python
# Put treatments into a list for for loop (and later for plot labels)
#set drugs to be analyzed, colors for the plots, and markers


# Create empty list to fill with tumor vol data (for plotting)
TumorVolList=[]
for drug in TragetDrugs:
    # Calculate the IQR and quantitatively determine if there are any potential outliers. 
    # Get Tumor Volume list for the drug
    FinalTumorvol=FinalTumorVolData[FinalTumorVolData['Drug Regimen']==drug]['Tumor Volume (mm3)'] 
    TumorVolList.append(FinalTumorvol) # Append the values to final list which contains tumor volumnes for all 4 drugs
    quartiles=FinalTumorvol.quantile([0.25,0.5,0.75])
    lowerq=quartiles[0.25]
    upperq=quartiles[0.75]
    iqr=upperq-lowerq
    lowerbound=lowerq-(1.5*iqr)
    upperbound=upperq+(1.5*iqr)
    outliers=FinalTumorvol.loc[(FinalTumorvol<lowerbound)|(FinalTumorvol>upperbound)] #Check for outliers
    if len(outliers)>0:
        print(f"Lower and Upper Bounds for drug regimen {drug} are {lowerbound} & {upperbound}")
        print(f"{drug} regimen's potential outliers :\n{outliers}\n'")
    else:
        print(f"Lower and Upper Bounds for drug regimen {drug} are {lowerbound} & {upperbound}")
        print(f"{drug} regimen does not have any outliers\n")
# Locate the rows which contain mice on each drug and get the tumor volumes

#Outlier Data for Infubinol
print("Outlier data for Infubinol")
pd.DataFrame(FinalTumorVolData.loc[15])
```

```python
# Determine outliers using upper and lower bounds
quartiles=FinalTumorVolData['Tumor Volume (mm3)'].quantile([0.25,0.5,0.75])
lowerq=quartiles[0.25]
upperq=quartiles[0.75]
iqr=upperq-lowerq
lowerbound=lowerq-(1.5*iqr)
upperbound=upperq+(1.5*iqr)
print(f"The lower quartile of Tumor Volume is: {lowerq}")
print(f"The upper quartile of Tumor Volume is: {upperq}")
print(f"The interquartile range of Tumor Volume is: {iqr}")
print(f"The the median of Tumor Volume is: {TargetDrugTumorVolume['Tumor Volume (mm3)'].median()}")
print(f"Values below {lowerbound} could be outliers.")
print(f"Values above {upperbound} could be outliers.")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
#establish x and y values and find St. Pearson Correlation Coefficient for Mouse Weight and Tumor Volume Avg
correlation=round(st.pearsonr(CapomulinData['Tumor Volume (mm3)'],CapomulinData['Weight (g)'])[0],2)

#print St. Pearson Correlation Coefficient
print(f"The correlation between both factors is {correlation}")

# establish linear regression values
model=st.linregress(CapomulinData['Weight (g)'],CapomulinData['Tumor Volume (mm3)'])
yvalues=CapomulinData['Weight (g)']*model[0]+model[1]
# linear regression line 
plt.plot(CapomulinData['Weight (g)'],yvalues,color='red')

# scatter plot of the data
plt.scatter(CapomulinData['Weight (g)'],CapomulinData['Tumor Volume (mm3)'])
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Tumore Volume Average')
plt.show()
```
